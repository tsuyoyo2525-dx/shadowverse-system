```python

Card_data = [
             {"name":"ダメージ魔法","cost":1,"hp":0,"atk":3,"efct":0,"deck":0,"order":0,"spell":1,"evo":0, "turn":0},
             {"name":"ドロー魔法",  "cost":1,"hp":0,"atk":0,"efct":0,"deck":0,"order":0,"spell":2,"evo":0, "turn":0},
             {"name":"トラ", "cost":1, "hp":1, "atk":1, "efct":2, "deck":0, "order":0, "spell":0, "evo":0, "turn":0},
             {"name":"キジ", "cost":2, "hp":3, "atk":1, "efct":1, "deck":0, "order":0, "spell":1, "evo":0, "turn":0},
             {"name":"サル", "cost":2, "hp":2, "atk":3, "efct":1, "deck":0, "order":0, "spell":1, "evo":0, "turn":0},
             {"name":"ブタ", "cost":3, "hp":1, "atk":5, "efct":2, "deck":0, "order":0, "spell":1, "evo":0, "turn":0},
             {"name":"リス", "cost":3, "hp":5, "atk":3, "efct":3, "deck":0, "order":1, "spell":1, "evo":0, "turn":0},
             {"name":"カバ", "cost":4, "hp":7, "atk":4, "efct":4, "deck":0, "order":1, "spell":1, "evo":0, "turn":0},
             {"name":"ウマ", "cost":5, "hp":5, "atk":5, "efct":0, "deck":0, "order":0, "spell":1, "evo":0, "turn":0},
             {"name":"クマ", "cost":6, "hp":9, "atk":9, "efct":0, "deck":0, "order":0, "spell":1, "evo":0, "turn":0},
            ]

import random


ahand = []
bhand = []
Astage = []
Bstage = []
Adeck = list(Card_data)
Bdeck = list(Card_data)

AHP = 20
BHP = 20
turn = 0
pp = 0
AEvo = 2
BEvo = 2
senkou = ""
play_check = True


def card_efct():
  # 使用カード（use）を選択 AAA
  use = int(input("攻撃するカードを選んでください（１～５）")) -1
  print("使用可能なカード:")
  for number, ch_card in enumerate(ahand):
    print(f"{number + 1}: {ch_card['name']} (ATK: {ch_card['atk']}, HP: {ch_card['hp']})")

  print("攻撃可能な対象:")
  for ch_number, ch_card in enumerate(Astage):
    if ch_card["efct"] == 3 or ch_card["efct"] == 4:
          print(f"{ch_number + 1}: {ch_card['name']} (ATK: {ch_card['atk']}, HP: {ch_card['hp']})")
    else:
      print(f"0: プレーヤー(HP: {AHP})")
      print(f"{ch_number + 1}: {ch_card['name']} (ATK: {ch_card['atk']}, HP: {ch_card['hp']})")


  # 使用カード（use）を選択 BBB
  print("使用可能なカード:")
  for number, ch_card in enumerate(bhand):
    print(f"{number + 1}: {ch_card['name']} (ATK: {ch_card['atk']}, HP: {ch_card['hp']})")

  print("攻撃可能な対象:")
  for ch_number, ch_card in enumerate(Bstage):
    if ch_card["efct"] == 3 or ch_card["efct"] == 4:
          print(f"{ch_number + 1}: {ch_card['name']} (ATK: {ch_card['atk']}, HP: {ch_card['hp']})")
    else:
      print(f"0: プレーヤー(HP: {BHP})")
      print(f"{ch_number + 1}: {ch_card['name']} (ATK: {ch_card['atk']}, HP: {ch_card['hp']})")
  choose = int(input("攻撃する対象の選択:"))

  #効果判定AAA
  if Astage[use]["efct"] == 0:
    print("効果なし")
  if Astage[use]["turn"] == 0:
    print("攻撃不可")
  elif Astage[use]["turn"] != 0:
    choose = int(input("攻撃する対象の番号(0～5)(プレーヤー0):"))

    while choose < 0 or len(Astage) < choose:    #有効な番号まで繰り返し
      print("無効な番号")
      choose = int(input("攻撃する対象の番号(0～5)(プレーヤー0)："))

    #マイターン
    #疾走カード
  elif Astage[use]["efct"] == 1:    #両方に攻撃可能確認
    choose = int(input("攻撃する対象の番号(0～5)(プレーヤー0)："))    #対象選択choose

    while choose < 0 or len(Astage) < choose:    #有効な番号まで繰り返し
      print("無効な番号")
      choose = int(input("攻撃する対象の番号(0～5)(プレーヤー0)："))

    if choose == 0:
      BHP -= Astage[use]["atk"]
    else:
      Bstage[choose]["hp"] -= Astage[use]["atk"]
  #突進カード
  elif Astage[use]["efct"] == 2 or Astage[use]["efct"] == 4:
    choose = int(input("攻撃する対象の番号(1～10)"))
    while choose < 1 or len(Bstage) < choose:
      print("無効な番号")
      choose = int(input("攻撃する対象の番号(1～10)"))



 #効果判定BBB
  if Bstage[use]["efct"] == 0:
    print("効果なし")
  elif Bstage[use]["turn"] == 0:
    print("攻撃不可")
  elif Bstage[use]["turn"] != 0:
    choose = int(input("攻撃する対象の番号w(0～5)(プレーヤー0):"))

    while choose < 0 or len(Astage) < choose:    #有効な番号まで繰り返し
      print("無効な番号")
      choose = int(input("攻撃する対象の番号(0～5)(プレーヤー0)："))

    #マイターン
    #疾走カード
  elif Bstage[use]["efct"] == 1:    #両方に攻撃可能確認
      choose = int(input("攻撃する対象の番号(0～5)(プレーヤー0)："))    #対象選択choose

      while choose < 0 or len(Astage) < choose:    #有効な番号まで繰り返し
        print("無効な番号")
        choose = int(input("攻撃する対象の番号(0～5)(プレーヤー0)："))

      if choose == 0:
        AHP -= Bstage[use]["atk"]
      else:
        Astage[choose]["hp"] -= Bstage[use]["atk"]
    #突進カード
  elif Bstage[use]["efct"] == 2:
    choose = int(input("攻撃する対象の番号(1～10)"))
    while choose < 1 or len(Astage) < choose:
      print("無効な番号")
      choose = int(input("攻撃する対象の番号(1～10)"))



#進化Evo
def Evolution():
  global AEvo, BEvo, Astage, Bstage, turn, Adeck, Bdeck
  Evo = input(("進化する/しない: Y / N"))
  if Evo == "Y":
    if turn % 2 != 0: # A's turn
      if AEvo <= 0:    #残り進化数確認
        print("進化不可")
      else:
        chHand = int(input("進化するカードを選択(左から1～5)")) - 1
        if 0 <= chHand < len(Astage) and Astage[chHand]["evo"] == 0: #進化前か確認
          AEvo -= 1
          Astage[chHand]["evo"] = 1    #evo 1なら進化済み
          Astage[chHand]["hp"] += 2    # hp + 2
          Astage[chHand]["atk"] += 2    # atk + 2
          print(f"{Astage[chHand]['name']}を進化しました")
        else:
          print("進化不可")
    else: # B's turn
      if BEvo <= 0:    #残り進化数確認
        print("進化不可")
      else:
        chHand = int(input("進化するカードを選択(左から1～5)")) - 1
        if 0 <= chHand < len(Bstage) and Bstage[chHand]["evo"] == 0: #進化前か確認
          BEvo -= 1
          Bstage[chHand]["evo"] = 1    #evo 1なら進化済み
          Bstage[chHand]["hp"] += 2    # hp + 2
          Bstage[chHand]["atk"] += 2    # atk + 2
          print(f"{Bstage[chHand]['name']}を進化しました")
        else:
          print("進化不可")


#スペルカード処理
def Spell():
  global ahand, bhand, Adeck, Bdeck, Astage, Bstage, turn,senkou
  if (senkou == "B" and turn%2 == 1) or (senkou == "A" and turn%2 == 0):
    for ch_card in bhand:
      if ch_card["spell"] == 1:    #ダメージスペル
        for i in reversed(range(len(Astage))):
          Astage[i]["hp"] -= 3
          if Astage[i]["hp"] <= 0:
            del Astage[i]
      if ch_card["spell"] == 2:    #カード追加スペル
        for i in range(2):
          if len(Bdeck) > 0:
            bhand.append(Bdeck.pop(0))
  else:
    for ch_card in ahand:
      if ch_card["spell"] == 1:    #ダメージスペル
        for i in reversed(range(len(Bstage))):
          Bstage[i]["hp"] -= 3
          if Bstage[i]["hp"] <= 0:
            del Bstage[i]
      if ch_card["spell"] == 2:    #カード追加スペル
        for i in range(2):
          if len(Adeck) > 0:
            ahand.append(Adeck.pop(0))





def Turnzero():
  global Adeck,Bdeck,turn,senkou
  if turn == 0:
    A = random.randint(0,1)
    if A == 0:
      senkou = "A"
      print("Aさんが先行です")
    else:
      senkou = "B"
      print("Bさんが先行です")
    random.shuffle(Adeck)
    random.shuffle(Bdeck)

def ahandchange():
  global Adeck,ahand,turn,A
  if turn == 0:
    ahand.extend(Adeck[0:4])
    Adeck = Adeck[5:]
    print(ahand)
    x = int(input("Aさん引き直す枚数を選択してください（０～４）"))
    if x > 0:
      redraw_indices = []
      for i in range(x):
        y = int(input(f"Aさん引き直したいカードを選んでください（１～{len(ahand)}）"))
        if 1 <= y <= len(ahand):
          redraw_indices.append(y - 1)
        else:
          print("無効な選択です。")
          return

      redraw_cards = [ahand[i] for i in sorted(redraw_indices, reverse=True)]
      for i in sorted(redraw_indices, reverse=True):
          ahand.pop(i)
      Adeck.extend(redraw_cards)
      random.shuffle(Adeck)

      new_cards = Adeck[0:len(redraw_indices)]
      ahand.extend(new_cards)
      Adeck = Adeck[len(redraw_indices):]
    print("新しい手札:", ahand)

def bhandchange():
  global Bdeck,bhand,turn,A
  if turn == 0:
    bhand.extend(Bdeck[0:4])
    Bdeck = Bdeck[5:]
    print(bhand)
    x = int(input("Bさん引き直す枚数を選択してください（０～４）"))
    if x > 0:
      redraw_indices = []
      for i in range(x):
        y = int(input(f"Bさん引き直したいカードを選んでください（１～{len(bhand)}）"))
        if 1 <= y <= len(bhand):
          redraw_indices.append(y - 1)
        else:
          print("無効な選択です。")
          return

      redraw_cards = [bhand[i] for i in sorted(redraw_indices, reverse=True)]
      for i in sorted(redraw_indices, reverse=True):
          bhand.pop(i)
      Bdeck.extend(redraw_cards)
      random.shuffle(Bdeck)

      new_cards = Bdeck[0:len(redraw_indices)]
      bhand.extend(new_cards)
      Bdeck = Bdeck[len(redraw_indices):]
    print("新しい手札:", bhand)


def Turn():
  global turn,Adeck,Bdeck,ahand,bhand,pp,senkou
  turn += 1
  for card in Astage:
      card["turn"] += 1
  for card in Bstage:
      card["turn"] += 1
  pp = (turn + 1)//2

  if (senkou == "B" and turn%2 == 1) or (senkou == "A" and turn%2 == 0):
    if len(bhand) < 10:
      if len(Bdeck) > 0:
        bhand.append(Bdeck.pop(0))
      print(bhand)
    else:
      if len(Bdeck) > 0:
        Bdeck.pop(0)

  else:
    if len(ahand) < 10:
      if len(Adeck) > 0:
        ahand.append(Adeck.pop(0))
      print(ahand)
    else:
      if len(Adeck) > 0:
        Adeck.pop(0)


def turnend():
  global play_check
  turnend = int(input("ターンを終了する場合は0,続ける場合は1を入力してください"))
  if turnend == 0:
    play_check = False
  else:
    play_check = True



def Battle():
  global Bstage,Astage,turn,AHP,BHP,senkou
  if (senkou == "B" and turn%2 == 1) or (senkou == "A" and turn%2 == 0):
    print("カードの番号は左から１～５です。プレイヤーを選ぶ場合は６を選択してください")
    for j in range(len(Bstage)):
      at = int(input("Bさんはどのカードで攻撃をするのかを選んでください（１～５）攻撃しない場合は０を入力")) - 1
      if 0 <= at < len(Bstage) and Bstage[at]["turn"] > 0:
        befct = [card["efct"] for card in Astage]
        if 3 not in befct and 4 not in befct:
          dfn = int(input("攻撃対象を選んでください（１～６）")) -1
          if dfn == 6:
            AHP  -= Bstage[at]["atk"]
          else:
            if 1 <= dfn <= len(Astage):
              dmg = Astage[dfn-1]["hp"] - Bstage[at]["atk"]
              Astage[dfn-1]["hp"] = dmg
              if Astage[dfn-1]["hp"] <= 0:
                Astage.pop(dfn-1)
            else:
              print("無効な対象です。")
        elif 3 in befct:
          defender_indices = [i for i, card in enumerate(Astage) if card["efct"] == 3]
          value = 1
          hyouji = [x + value for x in defender_indices]
          dfn = int(input(f"攻撃対象を選んでください{hyouji}")) - 1
          if dfn in defender_indices:
            dmg = Astage[dfn]["hp"] - Bstage[at]["atk"]
            Astage[dfn]["hp"] = dmg
            if Astage[dfn]["hp"] <= 0:
              Astage.pop(dfn)
          else:
            print("無効な対象です。")
        elif 4 in befct:
          defender_indices = [i for i, card in enumerate(Astage) if card["efct"] == 4]
          value = 1
          hyouji = [x + value for x in defender_indices]
          dfn = int(input(f"攻撃対象を選んでください{hyouji}")) - 1
          if dfn in defender_indices:
            dmg = Astage[dfn]["hp"] - Bstage[at]["atk"]
            Astage[dfn]["hp"] = dmg
            if Astage[dfn]["hp"] <= 0:
              Astage.pop(dfn)
          else:
            print("無効な対象です。")
      else:
        print("そのカードは攻撃できません。")


  else:
    print("カードの番号は左から１～５です。プレイヤーを選ぶ場合は６を選択してください")
    for j in range(len(Astage)):
      at = int(input("Aさんはどのカードで攻撃をするのかを選んでください（１～５）攻撃しない場合は０を入力")) - 1
      if 0 <= at < len(Astage) and Astage[at]["turn"] > 0:
        aefct = [card["efct"] for card in Bstage]
        if 3 not in aefct and 4 not in aefct:
          dfn = int(input("攻撃対象を選んでください（１～６）")) -1
          if dfn == 6:
            BHP  -= Astage[at]["atk"]
          else:
            if 1 <= dfn <= len(Bstage):
              dmg = Bstage[dfn-1]["hp"] - Astage[at]["atk"]
              Bstage[dfn-1]["hp"] = dmg
              if Bstage[dfn-1]["hp"] <= 0:
                Bstage.pop(dfn-1)
            else:
              print("無効な対象です。")
        elif 3 in aefct:
          defender_indices = [i for i, card in enumerate(Bstage) if card["efct"] == 3]
          value = 1
          hyouji = [x + value for x in defender_indices]
          dfn = int(input(f"攻撃対象を選んでください{hyouji}")) - 1
          if dfn in defender_indices:
            dmg = Bstage[dfn]["hp"] - Astage[at]["atk"]
            Bstage[dfn]["hp"] = dmg
            if Bstage[dfn]["hp"] <= 0:
              Bstage.pop(dfn)
          else:
            print("無効な対象です。")
        elif 4 in aefct:
          defender_indices = [i for i, card in enumerate(Bstage) if card["efct"] == 4]
          value = 1
          hyouji = [x + value for x in defender_indices]
          dfn = int(input(f"攻撃対象を選んでください{hyouji}")) - 1
          if dfn in defender_indices:
            dmg = Bstage[dfn]["hp"] - Astage[at]["atk"]
            Bstage[dfn]["hp"] = dmg
            if Bstage[dfn]["hp"] <= 0:
              Bstage.pop(dfn)
          else:
            print("無効な対象です。")
      else:
        print("そのカードは攻撃できません。")


def playCard():
  global turn,pp,ahand,bhand,Astage,Bstage,senkou
  if (senkou == "B" and turn%2 == 1) or (senkou == "A" and turn%2 == 0):
    while True:
      plcd = int(input("使いたいカードを選んでください（１～９）終了する場合は０を入力")) - 1
      if plcd == -1:
        break
      if 0 <= plcd < len(bhand) and pp >= bhand[plcd]["cost"]:
        Spell()
        Bstage.append(bhand[plcd])
        bhand.pop(plcd)
        pp -= Bstage[-1]["cost"]
        if Bstage[-1]["hp"] == 0:
          Bstage.pop(-1)
      elif 0 <= plcd < len(bhand) and pp < bhand[plcd]["cost"]:
        print("PPが足りません")
      else:
        print("無効な選択です")

    while len(Bstage) > 5:
      Bstage.pop(-1)

  else:
    while True:
      plcd = int(input("使いたいカードを選んでください（１～９）終了する場合は０を入力")) - 1
      if plcd == -1:
        break
      if 0 <= plcd < len(ahand) and pp >= ahand[plcd]["cost"]:
        Spell()
        Astage.append(ahand[plcd])
        ahand.pop(plcd)
        pp -= Astage[-1]["cost"]
        if Astage[-1]["hp"] == 0:
          Astage.pop(-1)
      elif 0 <= plcd < len(ahand) and pp < ahand[plcd]["cost"]:
        print("PPが足りません")
      else:
        print("無効な選択です")


    while len(Astage) > 5:
      Astage.pop(-1)


def showstage():
  global AHP,BHP,turn,AEvo,BEvo,pp,ahand,bhand,Astage,Bstage
  if (senkou == "B" and turn%2 == 1) or (senkou == "A" and turn%2 == 0):
    print("ーーーーBさんのターンですーーーー")
    print("あなたのHP：",BHP,"相手のHP：",AHP)
    print("現在のあなたの盤面",Bstage)
    print("相手の盤面：",Astage)
    print("使えるpp:",pp,"残り進化回数：",BEvo)
    print("あなたの手札：",bhand)
  else:
    print("ーーーーAさんのターンですーーーー")
    print("あなたのHP：",AHP,"相手のHP：",BHP)
    print("現在のあなたの盤面",Astage)
    print("相手の盤面：",Bstage)
    print("使えるpp:",pp,"残り進化回数：",AEvo)
    print("あなたの手札：",ahand)


def Result():
  global AHP,BHP,Adeck,Bdeck
  if AHP <= 0:
    print("Bさんの勝ちです")
    return False
  elif len(Adeck) <= 0:
    print("Bさんの勝ちです")
    return False
  elif BHP <= 0:
    print("Aさんの勝ちです")
    return False
  elif len(Bdeck) <= 0:
    print("Aさんの勝ちです")
    return False
  else:
    return True

def main():
  global Adeck,Bdeck,play_check
  Turnzero()
  ahandchange()
  bhandchange()
  while Result():
    Turn()
    play_check = True
    while play_check:
      showstage()
      playCard()
      Evolution()
      Battle()
      turnend()
main()
```
